```python
with open('export.csv', 'w' ,encoding="utf-8") as f:
    write = csv.writer(f)
    
    for pagenumber in range(1,10):
        r = session.get(f'https://api.digikala.com/v1/categories/mouse/search/?sort=7&seo_url=%2Fcategory-mouse%2F%3Fsort%3D7&page={pagenumber}')
        data_json = json.loads(r.text)
    
        for i in range(20):
            kala = data_json["data"]["products"][i]
            id = kala["id"]
    
            name = kala["title_fa"]
            all_star = kala["data_layer"]["metric6"]
            star_rate = kala["data_layer"]["dimension9"]
            price     = kala["default_variant"]["price"]["selling_price"]


            #################### feeling good rating
            try:
                total_count = kala["default_variant"]["statistics"]["total_count"]
            except:
                total_count = None
            try:
                total_rate = kala["default_variant"]["statistics"]["total_rate"]
            except:
                total_rate = None
            #########################  link
            link = f"https://www.digikala.com/product/dkp-{id}/"
            
            #########################  count comment
            comment = session.get(f'https://api.digikala.com/v1/product/{id}/comments/')
            comment = json.loads(comment.text)
            numc    = comment["data"]["pager"]["total_items"]
            
            #########################  count question
            question = session.get(f'https://api.digikala.com/v1/product/{id}/questions/')
            question = json.loads(question.text)
            numq     = question["data"]["pager"]["total_items"]
            row = [id,name,all_star,star_rate,price,total_count,total_rate,numc,numq,link]
            write.writerow(row)
```

```python
import pandas as pd 
columns = ["id","name","all_star","star_rate","price","total_count_feeling","total_rate_feeling","numc","numq","link"]

df = pd.read_csv("export.csv",names=columns)
df
```
