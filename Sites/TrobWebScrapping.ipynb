```python
from requests_html import HTMLSession
from selectolax.parser import HTMLParser

session = HTMLSession()
TARGET = "s23"
r = session.get(f'https://torob.com/search/?sort=popularity&size=80&query=s23query={TARGET}')
node = HTMLParser(r.text)


titles  = node.css(".product-name.jsx-2514672dc9197d80")
prices  = node.css(".product-price-text.jsx-2514672dc9197d80")
images  = node.css("img.jsx-221e9d776437f763")
shopes  = node.css(".shops.jsx-2514672dc9197d80")
links   = node.css(".cards.jsx-fadcb1a11849c1b4 a")


all_titles = [title.text() for title in titles]
all_prices = [price.text() for price in prices]
all_shopes = [shope.text() for shope in shopes]
image_urls = [image.attributes["srcset"].split(",")[1][:-3] for image in images]
all_links  = ["https://torob.com" + link.attributes["href"] for link in links]
all_links
```
