# Extract Data

```python
def extract_from_url(res_dic):
    a = res_dic
    lenght_include = a["included"]
    allinfo = []
    
    for i in range(len(lenght_include) - 1):
        vid = a["included"][i]["attributes"]["id"]
        title = a["included"][i]["attributes"]["title"]
        visit_cnt = int(a["included"][i]["attributes"]["visit_cnt_int"])
        big_poster	 = a["included"][i]["attributes"]["big_poster"]
        duration = int(a["included"][i]["attributes"]["duration"])
        date_exact_far	= a["included"][i]["attributes"]["date_exact"]
        date_exact_eng = a["included"][i]["attributes"]["sdate_rss"]
        like_cnt = int(a["included"][i]["attributes"]["like_cnt"])
        hd = a["included"][i]["attributes"]["hd"]
        
        ######### TODO
        # watch_watchTimeMinStr = a["included"][i]["attributes"]["watch"]["watchTimeMinStr"]
        # watch_durationPercentWatch = a["included"][i]["attributes"]["watch"]["durationPercentWatch"]
        # watch_avgWatchDuration = a["included"][i]["attributes"]["watch"]["avgWatchDuration"]
        # watch_text = a["included"][i]["attributes"]["watch"]["text"]
        
        item = [vid,title, visit_cnt, big_poster, duration, date_exact_far,date_exact_eng,like_cnt,hd]
        allinfo.append(item)
        
    return allinfo
```

# Export From APIs

```python
import requests
import pandas as pd 

USERID = "SeeTV"

url = f"https://www.aparat.com/api/fa/v1/user/video/list/username/{USERID}/"
df = pd.DataFrame()

for i in range(20):
    res = requests.get(url)
    json_res = res.json()
    data = extract_from_url(json_res) #ectract info
    df2 = pd.DataFrame(data, columns=["vid","title", "visit_cnt", "big_poster", "duration", "date_exact_far","date_exact_eng","like_cnt","hd"])
    df = pd.concat([df, df2], ignore_index=True)
    #extract next page
    try:
        url = json_res["data"][0]["attributes"]["link"]["next"]
    except:
        break 
```
