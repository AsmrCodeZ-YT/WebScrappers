```python
import requests
from pprint import pp
import pandas as pd
from pandas import json_normalize
import json



df = pd.DataFrame()

for i in range(7):

    num = i * 18
    header = {"User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:138.0) Gecko/20100101 Firefox/138.0"}
    url  = """https://www.compass.com/api/v3/agent_intelligence/agent/6089b3c45e8c6b00010d87e2/closed_deals?json={{"limit":18,"offset":{num},"orderBy":1,"closedDealsFilters":{{"closedDealType":0,"getHiddenDeals":false}}}}""".format(num=num)
    res = requests.get(url, headers=header)

    data = [i for i in json.loads(res.text)["listings"]]
    df1 = pd.json_normalize(data)
    df = pd.concat([df, df1], ignore_index=True)

df
df.to_csv("Export2.csv")
```

```python
from collections import Counter

all_tags = sum(df['listing.listingHashtags.hashtags'], [])
tag_freq = Counter(all_tags)

pp(dict(tag_freq.most_common(10)))
```
