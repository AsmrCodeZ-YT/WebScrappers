```python

from requests_html import HTMLSession
from selectolax.parser import HTMLParser
import pandas as pd
import re
import csv
```

```python
total_edit_popul_list = []
total_timesite_list   = []
total_timegame_list   = []
total_modelgame_list  = []
total_sizegame_list   = []
total_links_list      = []


for i in range(1,15):
    link_new = f"https://par30games.net/pc/page/{i}/"
    session = HTMLSession()
    s = session.get(link_new)
    node = HTMLParser(s.text)

    title_len     = node.css("main .posts .post span") #title and #timesite
    timegame_len  = node.css("main .posts .post ul li")
    links_list    = node.css("main .posts .post header h2 a")
    # try:
    popularity_list  = [title_len[n].text().split(" ")[-2] for n in range(len(title_len))]
    edit_popul_list  = []
    timesite_list    = [title_len[n].text().split(" ")[4:-2] for n in range(len(title_len))]
    timegame_list    = [timegame_len[n].text().split(": ")[1] for n in range(len(timegame_len))][::4]
    modelgame_list   = [timegame_len[n].text().split(": ")[1] for n in range(len(timegame_len))][1::4]
    sizegame_list    = [timegame_len[n].text().split(": ")[1] for n in range(len(timegame_len))][2::4]
    links_list       = [links_list[n].attributes["href"] for n in range(len(links_list))]

    for i in popularity_list:
        if "," in i:
            i = i.replace(",","")
        edit_popul_list.append(int(i))
        
    total_edit_popul_list += edit_popul_list 
    total_timesite_list   += timesite_list   
    total_timegame_list   += timegame_list   
    total_modelgame_list  += modelgame_list  
    total_sizegame_list   += sizegame_list   
    total_links_list      += links_list      
```

```python
df = pd.DataFrame(list(zip(total_edit_popul_list,
                      total_timesite_list,
                      total_timegame_list,
                      total_modelgame_list,
                      total_sizegame_list,
                      total_links_list)),
                  columns=["total_edit_popul_list","timegame_list","total_timesite_list",
                           "total_modelgame_list","total_sizegame_list","total_links_list"])
df
```

```python
df.to_csv("./info.csv")
```
