```python
from requests_html import HTMLSession
from selectolax.parser import HTMLParser

session = HTMLSession()
r = session.get(f'https://www.namasha.com/channel7117104301')
node = HTMLParser(r.text)


animes  = node.css(".mb-4, .my-4 .div div div")
links  = node.css(".flex-shrink-1")

all_anime = [anime.text().split() for anime in animes]
all_links = [link.attributes["href"] for link in links]

anime_info = []

for i in all_links:
    r = session.get(i)
    node = HTMLParser(r.text)

    likes     = node.css("html .video-like-btn .video-like-count")
    comments  = node.css("#comment-count")
    
    all_likes    = [like.text() for like in likes]
    all_comments = [comment.text() for comment in comments]
    
    anime_info.append([all_likes,all_comments])
    
######################### clean 
all_times = [i[0] for i in all_anime ]


######################### clean
newlikes = []
newcomment = []
# clean likes
for i in anime_info:
    newlikes.append(int(i[0][0]))
 
# clean comments   
for i in anime_info:
    if len(i[1]) == 0:
        newcomment.append(0)
    else :
        newcomment.append(int(i[1][0][1:-1]))
        
```

```python
import pandas as pd

df = pd.DataFrame(list(zip(all_links,
                           all_anime,
                           newlikes,
                           newcomment,
                           all_times)),
                  columns=["all_links","all_anime","newlikes","newcomment","all_times"])

df.sort_values("newcomment")
```
