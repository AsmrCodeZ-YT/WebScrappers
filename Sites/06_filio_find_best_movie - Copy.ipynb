```python
# list spliters 

def split_list(mylist, split_num):
    lenght = len(mylist)
    export_list = []
    temp_list = []
    count = 0
    
    if lenght % split_num == 0:
        for i in range(lenght):
            count += 1
            temp_list.append(mylist[i])
            if count == 3:
                export_list.append(temp_list)
                temp_list = []
                count = 0
    
    return export_list
# split times and views
def split_times_views(my_list):
    lenght = len(my_list)
    views = []
    times = []
    for i in my_list:
        for j,k in enumerate(i):
            if (j==0):
                views.append(k)
            if (j==2):
                times.append(k)
    return views,times

# create list
def create_links(links_list):
    temp = []
    for i in links_list:
        link = f"https://ifilo.net{i}"
        temp.append(link)
    return temp

# remove k: K to 1000
def remove_k(string, word):
    if word in string :
        return float(string[:-1])*1000
    else:
        return int(string)
```

```python
from requests_html import HTMLSession
from selectolax.parser import HTMLParser

PAGES = 10
Total_Titles = []
Total_Views  = []
Total_Times  = []
Total_Links  = []

for i in range(1,PAGES):
    url = f"https://ifilo.net/luffi/videos?p={i}"
    session = HTMLSession()
    r = session.get(url)
    node = HTMLParser(r.text)

    titles = node.css(".vb .vt a")
    likes = node.css(".vb .vi .du i")

    all_titles   = [title.text() for title in titles]
    all_links    = [title.attributes["href"] for title in titles]
    all_links    = create_links(all_links) # create list
    all_likes    = [like.text() for like in likes]
    all_likes    = split_list(all_likes,4)
    views ,times = split_times_views(all_likes)
    
    # append
    Total_Titles.append(all_titles)
    Total_Views.append(views)
    Total_Times.append(times)
    Total_Links.append(all_links)


Total_Titles = [element for nestedlist in Total_Titles for element in nestedlist]
Total_Views = [element for nestedlist in Total_Views for element in nestedlist]
Total_Times = [element for nestedlist in Total_Times for element in nestedlist]
Total_Links = [element for nestedlist in Total_Links for element in nestedlist]
```

```python
import pandas as pd

df = pd.DataFrame(list(zip(Total_Titles,
                           Total_Views,
                           Total_Times,
                           Total_Links)),
                  columns=["Total_Titles","Total_Views","Total_Times","Total_Links"])


df["Total_Views"] = df["Total_Views"].apply(lambda x: x.split()[0])
df["Total_Views"] = df["Total_Views"].apply(lambda x: remove_k(x, "K"))
df.sort_values("Total_Views",  ascending=False).head(15)
```
