# Single Page

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://soft98.ir/"
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

titles = [i.text.strip() for i in soup.find_all("h2", attrs={"class": "cbddt"})]
times = [i.text for i in soup.find_all("span", attrs={"class": "cbddid"})]
views = [int(i.text) for i in soup.find_all("span", attrs={"class": "cbddiv"})]
authors = [i.text for i in soup.find_all("span", attrs={"class": "cbddia"})]
categorical = [i.find("span").text for i in soup.find_all("div", attrs={"class": "cbddi cbddpi"})]

print(len(titles), len(times), len(views), len(authors), len(categorical))

df = pd.DataFrame(list(zip(titles, times, views, authors, categorical)),
               columns =['titles', 'times', "views", "authors", "categorical"])
```

# Multi Page

```python


PAGE_NUMBERS = 50
FILE_NAME = f"mutl_soft98_{PAGE_NUMBERS}.csv"

df = pd.DataFrame()

for pn in range(PAGE_NUMBERS):

    url = f"https://soft98.ir/page/{pn}/"
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')

    titles = [i.text.strip() for i in soup.find_all("h2", attrs={"class": "cbddt"})]
    times = [i.text for i in soup.find_all("span", attrs={"class": "cbddid"})]
    views = [int(i.text) for i in soup.find_all("span", attrs={"class": "cbddiv"})]
    authors = [i.text for i in soup.find_all("span", attrs={"class": "cbddia"})]
    categorical = [i.find("span").text for i in soup.find_all("div", attrs={"class": "cbddi cbddpi"})]

    print(len(titles), len(times), len(views), len(authors), len(categorical))
    df2 = pd.DataFrame(list(zip(titles, times, views, authors, categorical)),
               columns =['titles', 'times', "views", "authors", "categorical"])
    
    df = pd.concat([df, df2], ignore_index=True)

df.to_csv(FILE_NAME)
```
