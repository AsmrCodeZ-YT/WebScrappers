```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:129.0) Gecko/20100101 Firefox/129.0'}

url = 'https://store.playstation.com/en-us/concept/10009762'
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
response.status_code
```

```python
def try_extract(elment_name, attrs1, attrs2):
    try:
        
        print(soup.find(elment_name, attrs={attrs1: attrs2}).text)
    except:
        return None

def try_all_extract(elment_name, attrs1, attrs2):
    try:
        print([i.text for i in soup.find_all(elment_name, attrs={attrs1: attrs2})])
    except:
        return None

def try_all_extract_list(datalist):
    for i in datalist:
        try:
            print(soup.find("dd", attrs={"data-qa": i}).text)
        except:
           continue
```

```python
title = try_extract("h1", "data-qa", "mfe-game-title#name")
publisher = try_extract("div", "data-qa", 'mfe-game-title#publisher')
rating_from_5 = try_extract("div", "data-qa", "mfe-game-title#average-rating")
rating_count = try_extract("span", "data-qa","mfe-star-rating#overall-rating#total-ratings")
finalPrice = try_extract("span", "data-qa", "mfeCtaMain#offer0#finalPrice")
originalPrice = try_extract("span", "data-qa", "mfeCtaMain#offer0#originalPrice")
discountInfo = try_extract("span", "data-qa", "mfeCtaMain#offer0#discountInfo")
offer = try_extract("span", "class", "psw-t-overline psw-t-bold psw-l-line-left psw-fill-x")
discountDescriptor = try_extract("span", "data-qa", "mfeCtaMain#offer0#discountDescriptor")
release = try_extract("dd", "data-qa", "gameInfo#releaseInformation#releaseDate-value")
genre = try_extract("dd", "class", "gameInfo#releaseInformation#genre-value")
product_tag = try_all_extract("span", "class", "psw-p-x-2 psw-p-y-1 psw-t-tag")
noticesListItem3 = try_all_extract("li", "class", "psw-l-line-none psw-l-space-x-xs psw-l-shrink-wrap")
discribtionTag = try_all_extract("div", "class", "psw-l-switcher psw-with-dividers")
ratingProgresBar = try_all_extract("div", "class", "psw-rating-progress-bar psw-l-stack psw-fill-x")

alt_sub = ["gameInfo#releaseInformation#subtitles-value",
       "gameInfo#releaseInformation#ps5Subtitles-value",
       "gameInfo#releaseInformation#ps4Subtitles-value"]
subtitles = try_all_extract_list(alt_sub)

alt_voice = ["gameInfo#releaseInformation#Voice-value",
       "gameInfo#releaseInformation#ps5Voice-value",
       "gameInfo#releaseInformation#ps4Voice-value"]
voices = try_all_extract_list(alt_voice)
```
