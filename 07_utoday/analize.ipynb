# Frequency of Tags

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

df["tags"] = df["tags"].apply(lambda x: str(x))
all_words = ','.join(df['tags']).split(",")
word_count = Counter(all_words)


# for word, count in word_count.items():
#     print(f"{word}: {count}")
    
word_count_df = pd.DataFrame(word_count.items(), columns=['Tag', 'Count'])

word_count_df = word_count_df.sort_values(by='Count', ascending=False)[:20]

plt.figure(figsize=(10, 6))
plt.bar(word_count_df['Tag'], word_count_df['Count'], color='skyblue')
plt.xlabel('Tag')
plt.ylabel('Count')
plt.title('Frequency of Tags')
plt.xticks(rotation=90)
plt.show()
plt.savefig("Tags.csv")
```
