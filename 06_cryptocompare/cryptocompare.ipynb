# Single Page

```python
import pandas as pd
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import json


url = f"https://data-api.cryptocompare.com/asset/v1/top/list?page=1&page_size=100&sort_by=CIRCULATING_MKT_CAP_USD&sort_direction=DESC&groups=ID,BASIC,SUPPLY,PRICE,MKT_CAP,VOLUME,CHANGE,TOPLIST_RANK&toplist_quote_asset=USD&response_format=JSON"
req = requests.get(url)
print(req.status_code)

data = json.loads(req.text)
asset_list = []

for i in range(len(data["Data"]["LIST"])):
    
    asset_info = {
        "ID": data["Data"]["LIST"][i]["ID"],
        "SYMBOL": data["Data"]["LIST"][i]["SYMBOL"],
        "URI": data["Data"]["LIST"][i]["URI"],
        "ASSET_TYPE": data["Data"]["LIST"][i]["ASSET_TYPE"],
        "NAME": data["Data"]["LIST"][i]["NAME"],
        "LAUNCH_DATE": data["Data"]["LIST"][i].get("LAUNCH_DATE", None),
        "SUPPLY_MAX": data["Data"]["LIST"][i].get("SUPPLY_MAX", None),
        "SUPPLY_ISSUED": data["Data"]["LIST"][i].get("SUPPLY_ISSUED", None),
        "SUPPLY_CIRCULATING": data["Data"]["LIST"][i].get("SUPPLY_CIRCULATING", None),
        "SUPPLY_FUTURE": data["Data"]["LIST"][i].get("SUPPLY_FUTURE", None),
        "SUPPLY_LOCKED": data["Data"]["LIST"][i].get("SUPPLY_LOCKED", None),
        "SUPPLY_BURNT": data["Data"]["LIST"][i].get("SUPPLY_BURNT", None),
        "SUPPLY_STAKED": data["Data"]["LIST"][i].get("SUPPLY_STAKED", None),
        "PRICE_USD": data["Data"]["LIST"][i].get("PRICE_USD", None),
        "MKT_CAP_PENALTY": data["Data"]["LIST"][i].get("MKT_CAP_PENALTY", None),
        "CIRCULATING_MKT_CAP_USD": data["Data"]["LIST"][i].get("CIRCULATING_MKT_CAP_USD", None),
        "TOTAL_MKT_CAP_USD": data["Data"]["LIST"][i].get("TOTAL_MKT_CAP_USD", None),
        "SPOT_MOVING_24_HOUR_QUOTE_VOLUME_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_24_HOUR_QUOTE_VOLUME_USD", None),
        "SPOT_MOVING_24_HOUR_CHANGE_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_24_HOUR_CHANGE_USD", None),
        "SPOT_MOVING_24_HOUR_CHANGE_PERCENTAGE_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_24_HOUR_CHANGE_PERCENTAGE_USD", None),
        "SPOT_MOVING_7_DAY_QUOTE_VOLUME_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_7_DAY_QUOTE_VOLUME_USD", None),
        "SPOT_MOVING_7_DAY_CHANGE_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_7_DAY_CHANGE_USD", None),
        "SPOT_MOVING_7_DAY_CHANGE_PERCENTAGE_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_7_DAY_CHANGE_PERCENTAGE_USD", None),
        "SPOT_MOVING_30_DAY_QUOTE_VOLUME_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_30_DAY_QUOTE_VOLUME_USD", None),
        "SPOT_MOVING_30_DAY_CHANGE_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_30_DAY_CHANGE_USD", None),
        "SPOT_MOVING_30_DAY_CHANGE_PERCENTAGE_USD": data["Data"]["LIST"][i].get("SPOT_MOVING_30_DAY_CHANGE_PERCENTAGE_USD", None),
        "LOGO_URL": data["Data"]["LIST"][i].get("LOGO_URL", None),
    }
    asset_list.append(asset_info)



df = pd.DataFrame(asset_list)
df.to_csv('./data/crypto_assets.csv', index=False)
print("Data has been saved to 'crypto_assets.csv'")

```
